```python
import pandas as pd
from IPython import display
import shutil
```

```python
df = pd.read_csv("../dataset/signDatabasePublicFramesOnly/allAnnotations.csv", sep=";").loc[:, "Filename" : "Occluded,On another road"]
```

```python
df["Annotation tag"].value_counts()
```

```python
## SELECIONA APENAS AS PLACAS STOP E SPEED LIMIT
df_process = df.loc[(df["Annotation tag"] == "stop") | (df["Annotation tag"].str.contains("speedLimit"))]
```

```python
df_process["Annotation tag"].unique()
```

```python
 df_process.iloc[1]
```

```python
path = "../dataset/signDatabasePublicFramesOnly/"
img = path+"aiua120214-0/frameAnnotations-DataLog02142012_external_camera.avi_annotations/stop_1330545910.avi_image0.png"
img
```

```python
image = display.Image(img)
image
```

```python
## REMOVE AS PLACAS speedLimitUrdbl
df_process = df_process.loc[~(df["Annotation tag"] == "speedLimitUrdbl")]
```

```python
df_process = df_process.reset_index(drop=True)
```

```python
## RETIRA OS NUMEROS DAS PLACAS DE SPEED LIMIT
df_process["Annotation tag"] = df_process["Annotation tag"].str.replace("\d+", '')
```

```python
df_process["Annotation tag"].value_counts()
```

```python
## CORRIGE COLUNA OCCLUDED E REMOVE A COLUNA "Occluded,On another road"
df_process["Occluded"] = df_process["Occluded,On another road"].str.split(',').apply(lambda x : x[0])
df_process = df_process.drop("Occluded,On another road", axis=1)
```

```python
df_process["tag"] = df_process["Annotation tag"]
```

```python
df_process.loc[(df_process["tag"] == "stop"), "tag"] = 1
df_process.loc[(df_process["tag"] == "speedLimit"), "tag"] = 0
df_process
```

```python
df_to_train = df_process[["Filename", "Annotation tag", "tag"]]
```

```python
# !pip install Keras
# !pip install tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.layers.normalization import BatchNormalization
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from keras.preprocessing import image
from sklearn.model_selection import train_test_split
```

```python
classificador = Sequential()
classificador.add(Conv2D(32, (3,3), input_shape = (64, 64, 3), activation = 'relu'))
classificador.add(BatchNormalization())
classificador.add(MaxPooling2D(pool_size = (2,2)))
```

```python
classificador.add(Conv2D(32, (3,3), input_shape = (64, 64, 3), activation = 'relu'))
classificador.add(BatchNormalization())
classificador.add(MaxPooling2D(pool_size = (2,2)))
```

```python
classificador.add(Flatten())
```

```python
classificador.add(Dense(units = 128, activation = 'relu'))
classificador.add(Dropout(0.2))
classificador.add(Dense(units = 128, activation = 'relu'))
classificador.add(Dropout(0.2))
classificador.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
classificador.compile(optimizer = 'adam', loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
gerador_treinamento = ImageDataGenerator(rescale = 1./255,
                                         rotation_range = 7,
                                         horizontal_flip = True,
                                         shear_range = 0.2,
                                         height_shift_range = 0.07,
                                         zoom_range = 0.2)
```

```python
gerador_teste = ImageDataGenerator(rescale = 1./255)
```

```python
X = df_to_train.drop(columns=["tag"])
y = df_to_train["tag"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=12345)
```

```python
# COPIA OS ARQUIVOS PARA A PASTA DE TREINO E TEST
```

```python
train_folder = "../dataset/Train"
test_folder = "../dataset/Test"

# for filename, tag in zip(X_train["Filename"], X_train["Annotation tag"]):
# #     print(filename, tag)
#     shutil.copy(path+filename, train_folder+"/"+tag)
    
for filename, tag in zip(X_test["Filename"], X_test["Annotation tag"]):
    shutil.copy(path+filename, test_folder+"/"+tag)
```

```python
base_treinamento = gerador_treinamento.flow_from_directory('../dataset/Train', target_size = (64, 64), class_mode = 'binary')
```

```python
base_teste = gerador_teste.flow_from_directory('../dataset/Test', target_size = (64, 64), class_mode = 'binary')
```

```python
classificador.fit(base_treinamento, steps_per_epoch = 70, epochs = 10, validation_data = base_teste, validation_steps = 135)
```
